```python


# Introducci√≥n y explicaci√≥n:
print("""
Bienvenido al Juego M√°gico de Hechizos üßô‚Äç‚ôÇÔ∏è‚ú®

- Eres un mago enfrentando a un enemigo oscuro.
- Ambos inician con 100 puntos de salud.
- Puedes usar hechizos para atacar o defenderte.
- Cada hechizo tiene un n√∫mero limitado de usos, excepto el hechizo especial "revelar".
- "revelar" te muestra la salud tuya y del enemigo, y la cantidad de hechizos que te quedan.
- El hechizo "suerte" tiene efectos diferentes dependiendo de tu nombre de mago.
- El enemigo tambi√©n atacar√° dependiendo de tu nombre de mago.
- ¬°Derrota al enemigo antes de que √©l te derrote a ti!
""")

# Solicita el nombre del jugador
nombre_mago = input("¬øCu√°l es el nombre de tu mago?: ")
print(f"\n¬°Saludos, {nombre_mago}! Tu aventura comienza ahora.\n")

# Variables principales
salud_jugador = 100
salud_enemigo = 100

# Diccionario con los hechizos y sus usos permitidos (excepto 'revelar')
hechizos = {
    "fuego": 3,
    "curar": 3,
    "rayo": 2,
    "suerte": 2,
}

# Lista de efectos posibles del hechizo suerte simulando con el nombre del mago
efectos_suerte = ["curar", "x2", "contraataque"]
indice_suerte = 0  # control del uso basado en nombre
indice_ataque_enemigo = 0  # para controlar da√±o del enemigo con letras del nombre

# Funci√≥n para mostrar el estado actual
def mostrar_estado():
    print("\nEstado actual:")
    print(f"Salud de {nombre_mago}: {salud_jugador}")
    print(f"Salud del enemigo: {salud_enemigo}")
    print("Hechizos disponibles:")
    for h in hechizos:
        print("-", h, ":", hechizos[h], "uso(s) restante(s)")
    print("- revelar: ilimitado\n")

# Bucle principal
while salud_jugador > 0 and salud_enemigo > 0:
    hechizo = input("Escribe un hechizo (fuego, rayo, curar, suerte o revelar): ").lower()

    if hechizo == "revelar":
        mostrar_estado()
        continue  # no consume turno

    if hechizo not in hechizos:
        print("Hechizo no reconocido. Intenta de nuevo.\n")
        continue

    if hechizos[hechizo] == 0:
        print("Ya no puedes usar ese hechizo.\n")
        continue

    # Uso de hechizo v√°lido, se descuenta 1 uso
    hechizos_restantes = hechizos[hechizo]
    hechizos_restantes = hechizos_restantes - 1
    hechizos[hechizo] = hechizos_restantes  # Se guarda la nueva cantidad

    if hechizo == "fuego":
        print(f"üî• {nombre_mago} lanza una bola de fuego. El enemigo pierde 20 puntos.")
        salud_enemigo = salud_enemigo - 20

    elif hechizo == "rayo":
        print(f"‚ö° {nombre_mago} lanza un rayo. El enemigo pierde 30 puntos.")
        salud_enemigo = salud_enemigo - 30

    elif hechizo == "curar":
        print(f"üíö {nombre_mago} lanza un hechizo de curaci√≥n y recupera 15 puntos.")
        salud_jugador = salud_jugador + 15
        if salud_jugador > 100:
            salud_jugador = 100

    elif hechizo == "suerte":
        if indice_suerte < len(nombre_mago):
            letra = nombre_mago[indice_suerte].lower()
            indice_suerte += 1
        else:
            letra = "x"  # por defecto si ya no hay m√°s letras

        print(f"‚ú® Invocaste el hechizo de la SUERTE '...")

        if letra in ["a", "e", "i", "o", "u"]:
            print("ü©π ¬°La suerte te sonr√≠e! Recuperas 25 puntos de salud.")
            salud_jugador = salud_jugador + 25
            if salud_jugador > 100:
                salud_jugador = 100

        elif letra in ["x", "y", "z"]:
            print("üîÅ ¬°Contraataque! Devuelves 30 puntos de da√±o al enemigo.")
            salud_enemigo = salud_enemigo - 30

        elif indice_suerte % 2 == 0:
            print("üí• ¬°Da√±o doble activado! El enemigo pierde 40 puntos.")
            salud_enemigo = salud_enemigo - 40

        else:
            print("üîÅ ¬°Contraataque! Devuelves 30 puntos de da√±o al enemigo.")
            salud_enemigo = salud_enemigo - 30

    # Turno del enemigo (solo ataca si a√∫n est√° vivo)
    if salud_enemigo > 0:
        if indice_ataque_enemigo < len(nombre_mago):
            letra_enemigo = nombre_mago[indice_ataque_enemigo].lower()
            indice_ataque_enemigo += 1
        else:
            letra_enemigo = "x"  # por defecto

        if letra_enemigo in ["a", "e", "i", "o", "u"]:
            da√±o = 15
        elif letra_enemigo in ["l", "m", "n"]:
            da√±o = 30
        else:
            da√±o = 20

        print(f"\nüßü El enemigo lanza un ataque oscuro '. Pierdes {da√±o} puntos.")
        salud_jugador = salud_jugador - da√±o

    print("-"*40)  # Separador visual

# Resultado final
print("\nüéâ ¬°Fin del combate! üéâ")
if salud_jugador > 0:
    print(f"üèÜ {nombre_mago} ha derrotado al enemigo. ¬°Felicidades!")
else:
    print("‚ò†Ô∏è Has sido derrotado por el enemigo. Int√©ntalo de nuevo...")
```
