```python
# Import libraries
import pandas as pd

# Create a dataset
# Let´s build the values lists
value_n_list = ['Andrew', 'Phillip', 'John','Michael',
           'Laura', 'David', 'Sophie', 'Carlos',
           'Emma', 'Luis', 'Isabella']
value_w_list = [50,         78,      91.5, 85, 55, 88, 60, 72, 65, 80, 58] # [kg]
value_h_list = [1.76,      1.87,     1.69, 1.82, 1.60, 1.75, 1.65, 1.78, 1.68, 1.80, 1.62] # [m]

# Let´s use a dictionary ({key:value})
# with the format {"str":list}

data_dict = {"name":value_n_list,"weight":value_w_list,"height":value_h_list}

# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)
print(df)

# Preguntar cuántos registros desea agregar
cantidad = int(input("¿Cuántos registros desea agregar?: "))

for i in range(cantidad):
    print(f"\nRegistro {i+1} de {cantidad}")
    nombre = input("Ingrese el nombre: ")
    peso = float(input("Ingrese el peso (kg): "))
    altura = float(input("Ingrese la altura (m): "))

    # Agregar nuevo registro usando loc
    df.loc[len(df)] = [nombre, peso, altura]

    print("\nRegistro agregado correctamente.")

print("\nProceso finalizado. Datos completos:")
print(df)

```

```python
# Agregar columna edad
edad_lista = []
for i in range(len(df)):
    edad = int(input(f"Ingrese la edad de {df.loc[i, 'name']}: "))
    edad_lista.append(edad)

df['age'] = edad_lista

# Agregar columna calculada: IMC (Índice de Masa Corporal)
df['BMI'] = df['weight'] / (df['height'] ** 2)

# Agregar columna de clasificación del IMC
def clasificar_imc(bmi):
    if bmi < 18.5:
        return 'Bajo peso'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

df['BMI_Category'] = df['BMI'].apply(clasificar_imc)

print("\nProceso finalizado. Datos completos:")
print(df)
df
```
