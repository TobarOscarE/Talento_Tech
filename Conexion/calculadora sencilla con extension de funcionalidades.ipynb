```python
import math

print('Ejemplo Calculadora Sencilla (+,-,*,/) por Felipe Escallón:')
"""
ESTO ES UN COMENTARIO MULTILINEA
print('Ejemplo Calculadora Sencilla (+,-,*,/) por Felipe Escallón:\n')
PODEMOS DESCOMENTAR Y PROBAR EL print ANTERIOR
'''"""
print()
#MENU:
print("MENU:")
print("1: Sumar")
#print("1: +")
print("2: -")
print("3: *")
print("4: Dividir (/)")
print("5: Potenciacion")
print("6: Radicacion")
print("7: Logaritmacion")
print("8: Sin")
print("9: Cos")
print("10: Tan")
print("")



print()
#Ingreso de datos:
opcion=input("Ingrese la opción deseada del menú: ") # la opción se guarda como cadena
#opcion=int(input("Ingrese la opción deseada del menú: ")) # la opción se guarda como cadena
numero_uno=float(input('Ingrese el primer número de la operación:'))# guarda como flotante (número real)
numero_dos=float(input('Ingrese el segundo número de la operación:'))# guarda como flotante (número real)
#Posibilidades para ejecutar la calculadora:
#if opcion=='1':
#if opcion=='+':
#if opcion=='Sumar':
if opcion== "1":
  print("Vamos a sumar(+) los dos valores ingresados:")
  print('La suma es:',numero_uno + numero_dos)
if opcion=='2':
  print("Vamos a restar(-) los dos valores ingresados:\n")
  resta=numero_uno-numero_dos
  #print("Resultado="+str(resta)) # voy a concatenar dos cadenas(str1+str2: no suma sino que junta, str1str2)
  print("Resultado=",resta) # voy a concatenar dos cadenas(str1+str2: no suma sino que junta, str1str2)
if opcion=='3':
  print("Vamos a multiplicar(*) los dos valores ingresados:\n")
  #print(f'{numero_uno}*{numero_dos}={numero_uno*numero_dos}')
  mult=numero_uno*numero_dos
  print(f'{numero_uno}*{numero_dos}={mult}')
  #print('{}*{}'.format(numero_uno,numero_dos),"=",'{}'.format(mult))
if opcion=='4'or opcion=='/' or opcion=='Dividir':
  print("Vamos a dividir(/) los dos valores ingresados:\n")
  #VALIDANDO EL DENOMINADOR (debe ser diferente de cero)
  if numero_dos!=0:
    print(numero_uno/numero_dos)
  else:
    print("¡No es posible dividir entre cero!")
    
if opcion=='5'or opcion=="^" or opcion.lower()=='potenciacion':
  print(f"Vamos a potenciar {numero_uno} elevado a la {numero_dos} potencia:\n")
  print(f"{numero_uno} elevado a la {numero_dos} es: {numero_uno**numero_dos} ")

# opcion 6: Radicacion
if opcion == '6' or opcion.lower()== "radicacion":
    print(f"Vamos a hallar la raíz {numero_dos} de {numero_uno} usando potencias fraccionarias:")
    if numero_dos == 0:
        print("No se puede calcular una raíz de índice 0.")
    else:
        resultado = numero_uno ** (1 / numero_dos)
        print(f"La raíz {numero_dos} de {numero_uno} es: {resultado}")


# Opcion 7: Logaritmación
if opcion == '7' or opcion.lower() == 'logaritmacion' or opcion.lower() == "log":
    print(f"Vamos a calcular el logaritmo base 10 de {numero_uno} y {numero_dos} :")
    
    # Para numero_uno
    if numero_uno <= 0:
        print("El logaritmo no está definido para números menores o iguales a 0.")
    else:
        resultado_log1 = float(math.log10(numero_uno))
        print(f"log10({numero_uno}) = {resultado_log1} ")

    # Para numero_dos
    if numero_dos <= 0:
        print("El logaritmo no está definido para números menores o iguales a 0.")
    else:
        resultado_log2 = float(math.log10(numero_dos))
        print(f"log10({numero_dos}) = {resultado_log2}")

# Opción 8: Seno
if opcion == '8' or opcion.lower() == 'sin' or opcion.lower() == 'seno':
    print(f"Vamos a calcular el seno (sin) de {numero_uno}° y {numero_dos}°:")
    rad1 = math.radians(numero_uno)
    rad2 = math.radians(numero_dos)
    print(f"sin({numero_uno}°) = {math.sin(rad1)}")
    print(f"sin({numero_dos}°) = {math.sin(rad2)}")

# Opción 9: Coseno
if opcion == '9' or opcion.lower() == 'cos' or opcion.lower() == 'coseno':
    print(f"Vamos a calcular el coseno (cos) de {numero_uno}° y {numero_dos}°:")
    rad1 = math.radians(numero_uno)
    rad2 = math.radians(numero_dos)
    print(f"cos({numero_uno}°) = {math.cos(rad1)}")
    print(f"cos({numero_dos}°) = {math.cos(rad2)}")

# Opción 10: Tangente
if opcion == '10' or opcion.lower() == 'tan' or opcion.lower() == 'tangente':
    print(f"Vamos a calcular la tangente (tan) de {numero_uno}° y {numero_dos}°:")
    rad1 = math.radians(numero_uno)
    rad2 = math.radians(numero_dos)

    try:
        print(f"tan({numero_uno}°) = {math.tan(rad1)}")
    except:
        print(f"tan({numero_uno}°) no se puede calcular")

    try:
        print(f"tan({numero_dos}°) = {math.tan(rad2)}")
    except:
        print(f"tan({numero_dos}°) no se puede calcular")





```
